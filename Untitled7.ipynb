```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
```

```python
setosa_df = df[df['species'] == 'setosa']

X = setosa_df[['sepal_length']]
y = setosa_df['sepal_width']
```

```python
mean_sepal_length = X.mean()
mean_sepal_width = y.mean()

print(f"Rata-rata Sepal Length: {mean_sepal_length[0]}")
print(f"Rata-rata Sepal Width: {mean_sepal_width}")
```

```python
model = LinearRegression()

model.fit(X, y)

alpha = model.intercept_
beta = model.coef_[0]

print(f"Intercept (α): {alpha}")
print(f"Slope (β): {beta}")
```

```python
y_pred = model.predict(X)

plt.scatter(X, y, color='blue', label='Data Setosa')

plt.plot(X, y_pred, color='red', label='Garis Regresi')

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Regresi Linier Sepal Length vs Sepal Width (Setosa)')
plt.legend()

plt.show()
```
